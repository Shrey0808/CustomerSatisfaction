```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Used_bikes.csv")
```

```python
df
```

```python
# pieplot

brand = df['brand'].value_counts()

plt.pie(brand, labels=brand.index, autopct='%1.1f%%' , wedgeprops = {'ec':'k'} , shadow = True)
plt.title('Distribution of Bikes by Brand')
plt.show()
```

```python
# pieplot

owner = df['owner'].value_counts()

plt.pie(owner, labels=owner.index, autopct='%1.1f%%' , wedgeprops = {'ec':'k'} , shadow = True)
plt.title('Distribution of Bikes by ownership')
plt.show()
```

```python
# pieplot

age = df['age'].value_counts()

plt.pie(age, labels=age.index, autopct='%1.1f%%' , wedgeprops = {'ec':'k'} , shadow = True)
plt.title('Distribution of Bikes by age')
plt.show()
```

```python
# pieplot

city = df['city'].value_counts()

plt.pie(city, labels=city.index, autopct='%1.1f%%' , wedgeprops = {'ec':'k'} , shadow = True)
plt.title('Distribution of Bikes by City')
plt.show()
```

```python
# Box Plot
plt.boxplot(df['age'], vert = False)
plt.title('Box Plot of Bike by Age')
plt.xlabel('Age')
plt.show()
```

```python
age= df['age'].describe()

# Print the 5-number summary
print("5-Number Summary for Bike Age:")
print(f"Minimum: {age['min']}")
print(f"First Quartile (Q1): {age['25%']}")
print(f"Median: {age['50%']}")
print(f"Third Quartile (Q3): {age['75%']}")
print(f"Maximum: {age['max']}")
```

```python
# Box Plot
plt.boxplot(df['price'], vert = False)
plt.title('Box Plot of Bike by Price')
plt.xlabel('Price')
plt.show()
```

```python
price= df['price'].describe()

print("5-Number Summary for Bike Price:")
print(f"Minimum: {price['min']}")
print(f"First Quartile (Q1): {price['25%']}")
print(f"Median: {price['50%']}")
print(f"Third Quartile (Q3): {price['75%']}")
print(f"Maximum: {price['max']}")
```
