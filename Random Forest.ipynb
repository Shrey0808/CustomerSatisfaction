```python
import pandas as pd
```

```python
df = pd.read_csv("cust_satisfaction.csv")
df
```

```python
df.drop_duplicates(inplace = True)
```

```python
df = df.dropna()
```

```python
df3 = pd.get_dummies(df , drop_first = 'if_binary').astype('int') # One Hot Encoding Representation
df3.head()
```

```python
## Independent variable and dependent variables (X and Y)
x = df3.drop('satisfaction_satisfied', axis = 1)
x
```

```python
y = df3[['satisfaction_satisfied']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rdf = RandomForestClassifier()
```

```python
rdf.fit(x_train , y_train)
```

```python
# Model performance (model evaluation)
print("Training Score :",round(rdf.score(x_train,y_train),4)*100,"%")
print("Testing Score :",round(rdf.score(x_test,y_test),4)*100,"%")
```

```python
rdf2 = RandomForestClassifier(max_depth = 8, min_samples_split = 55 , min_samples_leaf = 25)
```

```python
rdf2.fit(x_train , y_train)
```

```python
# Model performance (model evaluation)
print("Training Score :",round(rdf2.score(x_train,y_train),4)*100,"%")
print("Testing Score :",round(rdf2.score(x_test,y_test['satisfaction_satisfied']),4)*100,"%")
```
